```python
import imdb
import pandas as pd
import numpy as np
```

```python
ia = imdb.IMDb()
dataset = pd.read_csv("data\movielens-20m-sample\movies.csv", header=0)
```

```python
keywords = []
for cnt, title in enumerate(dataset.title):
  search = ia.search_movie(title)
  if len(search)>0:
    movie = ia.get_movie(search[0].movieID, info="keywords")
    if "keywords" in movie:
      keywords.append("|".join(movie["keywords"][0:10]))
    else:
      print("No keywords for:",cnt,title)
      keywords.append("")
  else:
    print("did not find:",cnt,title)
    keywords.append("")
# print(title)
# print(ia.search_movie(title)[0].movieID)
# kw = ia.get_movie(ia.search_movie(title)[0].movieID, info="keywords")
```
